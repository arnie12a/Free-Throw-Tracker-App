```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import pandas as pd

# Initialize the WebDriver
driver = webdriver.Chrome()

url = "https://www.nba.com/stats/alltime-leaders?ActiveFlag=Yes&SeasonType=Regular%20Season&StatCategory=FTM"
driver.get(url)

# Wait for the table to load (you might need to adjust the waiting time or use WebDriverWait for more robust waiting)
driver.implicitly_wait(10)

# Find the table
table = driver.find_element(By.CLASS_NAME, "Crom_table__p1iZz")
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import pandas as pd

# Initialize the WebDriver
driver = webdriver.Chrome()

url = "https://www.nba.com/stats/alltime-leaders?ActiveFlag=Yes&StatCategory=FTM&SeasonType=Regular+Season"
driver.get(url)

# Wait for the table to load
wait = WebDriverWait(driver, 10)
wait.until(EC.presence_of_element_located((By.CLASS_NAME, "Crom_table__p1iZz")))

# Function to extract data from the current page
def extract_data():
    table = driver.find_element(By.CLASS_NAME, "Crom_table__p1iZz")
    headers = [th.text for th in table.find_elements(By.TAG_NAME, "th")]
    rows = []
    table_body = table.find_element(By.CLASS_NAME, "Crom_body__UYOcU")
    for tr in table_body.find_elements(By.TAG_NAME, "tr"):
        cells = tr.find_elements(By.TAG_NAME, "td")
        row = [cell.text.strip() for cell in cells]
        rows.append(row)
    return headers, rows

# Extract data from all pages
all_rows = []
while True:
    headers, rows = extract_data()
    all_rows.extend(rows)
    try:
        # Click the next page button
        next_button = driver.find_element(By.XPATH, "//a[@title='Next Page']")
        driver.execute_script("arguments[0].click();", next_button)
        # Wait for the next page to load
        wait.until(EC.staleness_of(table))
        wait.until(EC.presence_of_element_located((By.CLASS_NAME, "Crom_table__p1iZz")))
    except Exception as e:
        print(f"Error: {e}")
        break

# Close the WebDriver
driver.quit()

# Create a DataFrame
df = pd.DataFrame(all_rows, columns=headers)

# Save to CSV
df.to_csv("nba_alltime_leaders2.csv", index=False)

print("Data has been saved to nba_alltime_leaders.csv")

# Extract table headers
headers = [th.text for th in table.find_elements(By.TAG_NAME, "th")]

# Extract table rows
rows = []
table_body = table.find_element(By.CLASS_NAME, "Crom_body__UYOcU")
for tr in table_body.find_elements(By.TAG_NAME, "tr"):
    cells = tr.find_elements(By.TAG_NAME, "td")
    row = [cell.text.strip() for cell in cells]
    rows.append(row)

# Close the WebDriver
driver.quit()

# Create a DataFrame
df = pd.DataFrame(rows, columns=headers)

# Save to CSV
df.to_csv("nba_alltime_leaders2.csv", index=False)

print("Data has been saved to nba_alltime_leaders.csv")
```

```python
import pandas as pd
import json
df = pd.read_csv('nba_alltime_leaders.csv')
data = df[['PLAYER', 'FT%', 'FTM', 'FTA']].sort_values("FT%", ascending=False).drop_duplicates(subset=['PLAYER'])
data = data.reset_index()[['PLAYER', 'FT%','FTM', 'FTA']]
data_dict = data.to_dict(orient='records')
with open('players_ft.json', 'w') as json_file:
    json.dump(data_dict, json_file, indent=4)

```
